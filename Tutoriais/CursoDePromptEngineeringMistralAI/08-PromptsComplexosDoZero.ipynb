```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

```python
def gerar_resposta(prompt, max_tokens=600, temperature=0.3):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Criando um Prompt Complexo de Planejamento de Projeto

```python
prompt = (
    "Você é um gerente de projetos experiente. Crie um plano de projeto para desenvolver um aplicativo de controle financeiro pessoal.\n"
    "Formato da resposta:\n"
    "- Objetivo do Projeto: ...\n"
    "- Funcionalidades Principais: ...\n"
    "- Tecnologias Sugeridas: ...\n"
    "- Cronograma: ...\n"
    "- Riscos e Mitigações: ...\n"
    "Limite a resposta a 500 tokens."
)

print(gerar_resposta(prompt, max_tokens=500))
```

Exemplo 2: Simulando Diálogo com Restrições de Estilo e Personalidade

```python
prompt = (
    "Você é um personagem histórico: Albert Einstein. Responda à seguinte pergunta de forma espirituosa e informal:\n"
    "'O que você acha da inteligência artificial?'\n\n"
    "Formato da resposta:\n"
    "Einstein: ..."
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 3: Gerando um Relatório de Análise de Dados (Formato Estruturado)

```python
prompt = (
    "Você é um analista de dados. Gere um relatório de análise para o seguinte cenário:\n"
    "Cenário: A empresa XYZ teve uma queda de 15% nas vendas no último trimestre.\n\n"
    "Formato da resposta:\n"
    "- Causas Possíveis: ...\n"
    "- Impacto nos Resultados: ...\n"
    "- Ações Recomendadas: ...\n"
    "Limite a resposta a 400 tokens."
)

print(gerar_resposta(prompt, max_tokens=400))
```
