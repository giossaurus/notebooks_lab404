```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

```python
def gerar_resposta(prompt, max_tokens=400, temperature=0.0):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Prompt Sem Exemplos (Zero-Shot)

```python
prompt = "Traduza a frase 'I am happy' para o francês."
print(gerar_resposta(prompt, max_tokens=100))
```

Exemplo 2: Prompt com Exemplos (Few-Shot)


```python
prompt = (
    "Aqui estão alguns exemplos de tradução de inglês para francês:\n"
    "1. 'Good morning' -> 'Bonjour'\n"
    "2. 'Thank you' -> 'Merci'\n"
    "3. 'How are you?' -> 'Comment ça va?'\n\n"
    "Agora traduza a seguinte frase:\n"
    "'I am happy' ->"
)

print(gerar_resposta(prompt, max_tokens=100))
```

Exemplo 3: Few-Shot para Classificação de Sentimentos

```python
prompt = (
    "Classifique o sentimento das seguintes frases como 'Positivo', 'Negativo' ou 'Neutro'.\n"
    "Exemplos:\n"
    "1. 'Eu adorei o novo filme.' -> Positivo\n"
    "2. 'Estou muito triste com o resultado.' -> Negativo\n"
    "3. 'Foi um dia comum.' -> Neutro\n\n"
    "Agora classifique:\n"
    "'A comida estava deliciosa.' ->"
)

print(gerar_resposta(prompt, max_tokens=100))
```
