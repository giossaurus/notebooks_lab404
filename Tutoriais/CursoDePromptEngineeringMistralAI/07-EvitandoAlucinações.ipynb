```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

```python
def gerar_resposta(prompt, max_tokens=400, temperature=0.0):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Pergunta Ambígua (Modelo pode alucinar)

```python
prompt = "Quantos metros de altura tem o castelo do Harry Potter na vida real?"
print(gerar_resposta(prompt, max_tokens=200))
```

Exemplo 2: Forçando o Modelo a Admitir que Não Sabe (Melhor prática)

```python
prompt = (
    "Se você não tiver certeza absoluta da resposta, diga claramente 'Não sei'.\n"
    "Pergunta: Quantos metros de altura tem o castelo do Harry Potter na vida real?"
)

print(gerar_resposta(prompt, max_tokens=200))
```

Exemplo 3: Pedindo Fontes ou Referências

```python
prompt = (
    "Forneça a resposta apenas se puder citar uma fonte oficial confiável. "
    "Pergunta: Qual é a distância exata entre a Terra e Marte?"
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 4: Confirmando Respostas com 'True/False'

```python
prompt = (
    "A seguinte afirmação é verdadeira ou falsa?\n"
    "'Os pinguins vivem no Polo Norte.'\n"
    "Responda apenas com 'Verdadeiro' ou 'Falso' e uma breve explicação."
)

print(gerar_resposta(prompt, max_tokens=150))
```
