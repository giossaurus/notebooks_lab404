Configurações da API:

```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

 Função para gerar resposta e exibir tokens usados:

```python
def gerar_resposta(prompt, max_tokens=300, temperature=0.0):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Dados Misturados com Instrução (Ruim)

```python
prompt = (
    "Nome: João\nIdade: 28\nCargo: Analista de Dados\n"
    "Faça uma apresentação de João para uma reunião de equipe."
)
print(gerar_resposta(prompt, max_tokens=200))
```

Exemplo 2: Separação Clara de Dados e Instrução (Melhor Prática)

```python
prompt = (
    "## Dados\n"
    "Nome: João\nIdade: 28 anos\nCargo: Analista de Dados\nExperiência: 5 anos em projetos de BI\n\n"
    "## Instrução\n"
    "Crie uma breve apresentação profissional de João para uma reunião de equipe, "
    "enfatizando suas habilidades em projetos de Business Intelligence."
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 3: Usando Blocos Delimitadores (Formato Avançado)

```python
prompt = (
    "### DADOS\n"
    "Produto: Curso de Engenharia de Prompts\n"
    "Duração: 8 semanas\n"
    "Público-alvo: Desenvolvedores e Analistas de Dados\n\n"
    "### INSTRUÇÃO\n"
    "Crie um texto promocional para o produto acima, destacando a duração e o público-alvo. "
    "Limite a resposta a 3 parágrafos curtos."
)

print(gerar_resposta(prompt, max_tokens=300))
```
