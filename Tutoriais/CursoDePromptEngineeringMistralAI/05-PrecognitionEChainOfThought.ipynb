```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

```python
def gerar_resposta(prompt, max_tokens=300, temperature=0.0):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Sem Raciocínio Intermediário (Resposta direta)

```python
prompt = "Qual é o resultado de 18 * 12?"
print(gerar_resposta(prompt, max_tokens=100))
```

Exemplo 2: Forçando Raciocínio Passo a Passo (Chain of Thought)

```python
prompt = (
    "Vamos pensar passo a passo.\n"
    "Qual é o resultado de 18 * 12? "
    "Explique cada etapa do cálculo antes de apresentar a resposta final."
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 3: Resolvendo um Problema Complexo Passo a Passo

```python
prompt = (
    "Você é um planejador financeiro. Um cliente tem uma dívida de R$ 5.000, "
    "com juros de 2% ao mês. Ele pode pagar R$ 500 por mês. "
    "Calcule quantos meses ele levará para quitar a dívida e mostre os cálculos passo a passo."
)

print(gerar_resposta(prompt, max_tokens=300))
```
