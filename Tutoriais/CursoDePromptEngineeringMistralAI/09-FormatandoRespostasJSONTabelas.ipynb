```python
import os
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "sua_chave_de_api_aqui")
client = Mistral(api_key=api_key)
MODEL_NAME = "mistral-small-latest"
```

```python
def gerar_resposta(prompt, max_tokens=600, temperature=0.0):
    response = client.chat.complete(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=temperature
    )
    print(f"Tokens usados: {response.usage.total_tokens}")
    return response.choices[0].message.content
```

Exemplo 1: Retornando Resposta no Formato JSON

```python
prompt = (
    "Forneça um resumo das vendas trimestrais no seguinte formato JSON:\n"
    "{\n"
    '  "total_vendas": <número>,\n'
    '  "produto_mais_vendido": "<nome do produto>",\n'
    '  "regiao_com_maior_crescimento": "<região>"\n'
    "}\n"
    "Utilize valores fictícios."
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 2: Retornando uma Tabela em Markdown

```python
prompt = (
    "Liste os 5 principais países por PIB em 2023 no formato de tabela Markdown.\n"
    "Formato esperado:\n"
    "| País       | PIB (USD Trilhões) |\n"
    "|------------|--------------------|\n"
    "| Exemplo    | 1.23               |\n"
)

print(gerar_resposta(prompt, max_tokens=300))
```

Exemplo 3: Tabela de Comparação de Modelos de IA

```python
prompt = (
    "Crie uma tabela comparando os seguintes modelos de IA: GPT-4, Claude 3, Mistral 7B, Llama 2.\n"
    "A tabela deve incluir as colunas: Modelo, Tamanho (B), Open Source (Sim/Não), Melhor Uso.\n"
    "Formato esperado:\n"
    "| Modelo      | Tamanho (B) | Open Source | Melhor Uso          |\n"
    "|-------------|-------------|--------------|---------------------|\n"
)

print(gerar_resposta(prompt, max_tokens=400))
```
